# Work on iris dataset in seaborn

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.lineplot(x="sepal_length",y="petal_length",data=iris)
```

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.barplot(x="sepal_length",y="petal_length",data=iris)
```

```python
sns.boxplot(x="sepal_length",y="species",data=iris)
```

```python
iris.describe()
```

```python
sns.catplot(x="sepal_length",y="petal_length",hue="species",data=iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(),annot=True)
```
